# Numerical Data Visualization : 수치형 데이터 

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic
```

## 1. histplot
- 히스토그램(histogram)은 데이터의 구간별 빈도수를 나타내는 그래프

```python
sns.histplot(data=titanic, x='age');
```

```python
sns.histplot(data=titanic, x='age', bins=10);
```

```python
sns.histplot(data=titanic, x='age', hue='alive');
```

```python
sns.histplot(data=titanic, x='age', hue='alive', multiple='stack');
```

## 2. kdeplot
- 커널밀도추정(kernel dentisty estimation) 함수는 히스토그램을 매끄럽게 곡선으로 연결한 그래프로 이해

```python
sns.kdeplot(data=titanic, x='age', hue='alive', multiple='stack');
```

## 3. displot
- 수치형 데이터의 분포를 나타내는 그래프
- histplot()과 kdeplot()을 동시에 그릴 수 있음

```python
sns.displot(data=titanic, x='age');
```

```python
sns.displot(data=titanic, x='age', kind='kde');
```

```python
sns.displot(data=titanic, x='age', kde=True);
```

## 4. rugplot
- 주변 분포(marginal distribution)를 나타내는 그래프
- 단독으로 사용하기 보다는 주로 다른 분포도 그래프와 함께 사용
- 단일 피처가 어떻게 분포되어 있는지를 x축 위에 작은 선분(rug)으로 표시

```python
sns.kdeplot(data=titanic, x='age')
sns.rugplot(data=titanic, x='age');
```

# Categorical Data Visualization : 범주형 데이터 시각화

## 1. barplot
- 범주형 데이터에 따른 수치형 데이터의 평균과 신뢰구간을 그려준다.
- 수치형 데이터의 평균은 막대 높이로, 신뢰구간은 오차 막대로 표현한다.
- 원본 데이터를 복원 샘플링하여 얻은 표본을 활용해 평균과 신뢰구간을 구한다.
- 즉 원본 데이터 평균이 아니라 샘플링한 데이터 평균을 구한다.
- x 파라미터에 범주형 데이터를, y 파라미터에 수치형 데이터를 전달한다.

```python
sns.barplot(x='class', y='fare', data=titanic);
```

## 2. pointplot
- 막대그래프와 모양만 다를 뿐 동일한 정보를 제공

```python
sns.pointplot(x='class', y='fare', data=titanic);
```

## 3. boxplot

```python
sns.boxplot(x='class', y='age', data=titanic);
```

```python
sns.boxplot(x='class', y='age', hue='sex', data=titanic);
```

## 4. violinplot
- 박스플롯과 커널밀도추정 함수 그래프를 합쳐놓은 그래프로 볼 수 있음

```python
sns.violinplot(x='class', y='age', data=titanic);
```

```python
sns.violinplot(x='class', y='age', hue='sex', data=titanic, split=True);
```

## 5. countplot
- 범주형 데이터의 개수를 확인할 때 사용

```python
sns.countplot(x='class', data=titanic);
```

```python
sns.countplot(y='class', data=titanic);
```

## barplot() vs countplot()
- barplot()은 범주형 데이터별 수치형 데이터의 평균을 구해주기 때문에 피처를 2개 받는다
- countplot()은 피처를 범주형 데이터 하나만 받는다
- barplot()으로는 평균이 아닌 중앙값, 최댓값, 최솟값을 구할 수도 있다.

```python
sns.barplot(x='class', y='fare', data=titanic);
```

```python
sns.countplot(y='class', data=titanic);
```

```python
import numpy as np
sns.barplot(x='class', y='fare', data=titanic, estimator=np.median);
```

```python
sns.barplot(x='class', y='fare', data=titanic, estimator=np.max);
```

```python
sns.barplot(x='class', y='fare', data=titanic, estimator=np.min);
```

## 6. pie

```python
import matplotlib.pyplot as plt

x = [10, 60, 30]
labels = ['A', 'B', 'C']

plt.pie(x=x, labels=labels, autopct='%.1f%%');
```

# Data Relation Visualization : 데이터 관계 시각화

## 1. heatmap

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

flights_pivot
```

```python
sns.heatmap(data=flights_pivot);
```

## 2. lineplot
- x 파라미터에 전달한 값에 따라 y 파라미터에 전달한 값의 평균과 95% 신뢰구간을 나타낸다.

```python
sns.lineplot(x='year', y='passengers', data=flights);
```

## 3. scatterplot

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips);
```

```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=tips);
```

## 4. regplot
- 산점도와 선형 회귀선을 동시에 그려주는 함수

```python
sns.regplot(x='total_bill', y='tip', data=tips);
```

```python
sns.regplot(x='total_bill', y='tip', data=tips, ci=99);
```
